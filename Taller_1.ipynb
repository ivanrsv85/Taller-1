El top 10 de artistas más activos de los últimos 10 años, usando la columna *energy* como medida de actividad (según su definición) es:

```python
#spotify_df.query('year'>2012).groupby('artist')['energy'].sum()
# current date minus a year
from datetime import datetime
from dateutil.relativedelta import relativedelta

# minus number of year
currentTimeDate = datetime.now() - relativedelta(years=10)
tenYearsAgo = int(currentTimeDate.strftime('%Y'))


#top_artist = spotify_df[spotify_df['year'] >= tenYearsAgo].groupby('artist').agg({'energy':sum}).sort_values(by='energy')
spotify_df[spotify_df['year'] >= tenYearsAgo].groupby('artist').agg({'energy':sum, }).sort_values(by='energy', ascending=False).head(10)
```
